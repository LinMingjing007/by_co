## 2. 导入依赖

本案例中，图像风格化主流程由 OpenCV 和 MindSpore 完成；风格提示词解析优先接入 MindNLP 分词组件。


```python
import cv2
import mindnlp
import mindspore
from photo2style.pipeline import Photo2StylePipeline, parse_style_text

print('MindSpore 版本:', mindspore.__version__)
print('MindNLP 版本:', mindnlp.__version__)
```

## 3. 初始化推理管线

推理管线支持 MindSpore 后端；若环境中未启用对应组件，也可以回退到 NumPy 路径。


```python
pipe = Photo2StylePipeline(use_mindspore=True, device_target='CPU')
```

## 4. 解析风格提示词

这里演示如何用中文提示词控制生成风格。解析函数会优先尝试使用 MindNLP 分词器做标准化切分。


```python
style_name, message = parse_style_text('请生成吉卜力动画风格')
print(style_name)
print(message)
```

## 5. 执行单图推理

请将 `input.jpg` 替换为你的真人照片路径。


```python
input_path = 'input.jpg'
output_path = 'outputs/out.jpg'

result = pipe.stylize_path(input_path, style_name=style_name)
output_bgr = cv2.cvtColor(result.image_rgb, cv2.COLOR_RGB2BGR)
Path(output_path).parent.mkdir(parents=True, exist_ok=True)
cv2.imwrite(output_path, output_bgr)
print('输出已保存到:', output_path)
```
